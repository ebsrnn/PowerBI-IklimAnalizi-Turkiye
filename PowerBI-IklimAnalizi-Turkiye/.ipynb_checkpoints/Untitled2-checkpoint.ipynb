```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

class Weather:
    def __init__(self):
        self.url = "https://www.netdata.com/netsite/2e012549/sehir-koordinatlari?p="
        self.browser = webdriver.Firefox()
        self.data = []

    def cityWeather(self):
        page = 1
        while True:
            print(f"Sayfa: {page}")
            self.browser.get(self.url + str(page))
            time.sleep(2)

            rows = self.browser.find_elements(By.XPATH, "//*[@id='no-more-tables']/table/tbody/tr")
            if not rows:
                print("Veri bulunamadı, işlem tamam.")
                break

            for row in rows:
                try:
                    cells = row.find_elements(By.TAG_NAME, "td")
                    if len(cells) >= 5:
                        il = cells[2].text.strip()
                        enlem = cells[3].text.strip()
                        boylam = cells[4].text.strip()

                        self.data.append({
                            "il": il,
                            "enlem": enlem,
                            "boylam": boylam
                        })
                except Exception as e:
                    print("Satır hatası:", e)
                    continue

            page += 1

        self.browser.quit()
        df = pd.DataFrame(self.data)
        df.to_csv("sehir_koordinatlari.csv", index=False)
        print(f"Tüm veriler kaydedildi. Toplam {len(self.data)} kayıt.")

# Çalıştır
weather = Weather()
weather.cityWeather()
```

```python
import time
import csv
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.firefox.options import Options

def veri_topla_ve_kaydet():
    # Firefox gizli mod ayarı
    options = Options()
    options.add_argument("-private")
    driver = webdriver.Firefox(options=options)

    # Yıllar ve sektörler listesi
    yillar = [2021, 2022, 2023, 2024]
    sektorler = ["Power", "Waste"]

    # Admin kodları
    admin_kodlari = {
        "ülke": "Turkey%20(TUR):40071:TUR:country",
        "şehir": "Istanbul%20Province--%20TUR:2163:TUR.40_1:state",
        "ilçe": "Ey%C3%BCp%20District--%20Istanbul%20Province--%20TUR:29470:TUR.40.19_1:county"
    }

    # URL şablonu
    url_sablonu = "https://climatetrace.org/explore/#admin={admin}&gas=ch4&year={year}&timeframe=100&sector={sector}&asset="

    # Sonuçların tutulacağı liste
    tum_sonuclar = []
    alanlar = ["Yıl", "Ülke", "Şehir", "İlçe"] + sektorler

    for yil in yillar:
        satir = {"Yıl": yil}
        
        # Ülke, şehir ve ilçe bilgilerini doldur
        for seviye, admin_kodu in admin_kodlari.items():
            if seviye == "ülke":
                satir["Ülke"] = admin_kodu.split(":")[0].replace("%20", " ")
            elif seviye == "şehir":
                satir["Şehir"] = admin_kodu.split(":")[0].replace("%20", " ").replace("--", "-")
            elif seviye == "ilçe":
                satir["İlçe"] = admin_kodu.split(":")[0].replace("%20", " ").replace("--", "-").replace("%C3%BC", "ü")

        for sektor in sektorler:
            try:
                admin_kodu = admin_kodlari["ilçe"]
                url = url_sablonu.format(admin=admin_kodu, year=yil, sector=sektor)

                print(f"\n🔗 Yıl: {yil}, Sektör: {sektor} → URL açılıyor...")
                driver.get(url)
                time.sleep(8)

                # Sayfa yüklendikten sonra yenileme
                driver.refresh()
                print("♻️ Sayfa yenilendi.")
                time.sleep(6)

                try:
                    veri_elementi = driver.find_element(By.XPATH, "/html/body/div[4]/div[2]/div[2]")
                    veri = veri_elementi.text.strip()
                    print(f"✅ {sektor} verisi: {veri}")
                except Exception:
                    veri = "Veri bulunamadı"
                    print(f"⚠️ {sektor} verisi alınamadı.")

                satir[sektor] = veri

            except Exception as e:
                print(f"❌ Hata oluştu: {e}")
                satir[sektor] = "Hata"

            # Her sektör sonrası kısa bekleme
            time.sleep(3)

        tum_sonuclar.append(satir)

    driver.quit()

    # CSV'ye yaz
    with open("climatetrace_duzenli.csv", mode="w", newline="", encoding="utf-8") as dosya:
        yazici = csv.DictWriter(dosya, fieldnames=alanlar)
        yazici.writeheader()
        yazici.writerows(tum_sonuclar)

    print("\n✅ Tüm veriler climatetrace_duzenli.csv dosyasına kaydedildi.")

# Fonksiyonu çalıştır
veri_topla_ve_kaydet()
```
